```python
def sound2img(filename, category):
    # Load audio
    wav, sr = librosa.load('ESC-50-master/audio/'+ filename, sr=None)
    librosa.display.waveplot(wav, sr)
    plt.colorbar()
    plt.close()

    # Generate spectogram
    X = librosa.stft(wav)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()
    plt.close()

    filen = filename.replace("wav","png")

    # replace filename
    fig = plt.figure()
    plt.axis('off')
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
    fig.savefig('D:/proyectos/sound/images/{}/{}'.format(str(category), filen))
    plt.close()
```

```python
for index, row in metadata.iterrows():
    sound2img(row["filename"], row["category"])
    
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("acc_loss.csv")
```

```python
df.head()
```

```python
def plot_accuracies(df):
    accuracies = df['val_acc']
    plt.plot(accuracies, '-x')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.title('Accuracy vs. No. of epochs');
```

```python
def plot_losses(df):
    train_losses = df['train_loss']
    val_losses = df['val_loss']
    plt.plot(train_losses, '-bx')
    plt.plot(val_losses, '-rx')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend(['Training', 'Validation'])
    plt.title('Loss vs. No. of epochs');
```

```python
plot_accuracies(df)
```

```python
plot_losses(df)
```
